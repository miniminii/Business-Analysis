```python
from app_store_scraper import AppStore
from pprint import pprint
```

```python
abc_mart = AppStore(country="kr", app_name="ABC-MART")
abc_mart.review(how_many=1000) #최대 개수 지정

pprint(abc_mart.reviews)
pprint(abc_mart.reviews_count)
```

```python
abc_mart = pd.DataFrame(abc_mart.reviews)
print(len(abc_mart))
abc_mart
```

```python
art = AppStore(country="kr", app_name="A-RT.COM")
art.review(how_many=1000)

pprint(art.reviews)
pprint(art.reviews_count)
```

```python
art = pd.DataFrame(art.reviews)
print(len(art))
art
```

```python
grand_stage = AppStore(country="kr", app_name="GRAND STAGE")
grand_stage.review(how_many=1000)

pprint(grand_stage.reviews)
pprint(grand_stage.reviews_count)
```

```python
grand_stage = pd.DataFrame(grand_stage.reviews)
print(len(grand_stage))
grand_stage
```

```python
on_the_spot = AppStore(country="kr", app_name="ON THE SPOT")
on_the_spot.review(how_many=1000)

pprint(on_the_spot.reviews)
pprint(on_the_spot.reviews_count)
```

```python
on_the_spot = pd.DataFrame(on_the_spot.reviews)
print(len(on_the_spot))
on_the_spot
```

```python
merge = pd.concat([abc_mart, art, grand_stage, on_the_spot], axis=0)
merge
```

```python
merge.to_csv("abc_review.csv", index=False)
```
